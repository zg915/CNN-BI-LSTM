```python
import pandas as pd
```

```python
sp_df = pd.read_csv('SP500_raw.csv')
sp_df = sp_df.drop(['Adj Close'],axis=1)
```

```python
sp_df['Average Price'] = (sp_df['High'] + sp_df['Low']) / 2
sp_df['Turnover'] = sp_df['Volume'] * sp_df['Average Price']
sp_df = sp_df.drop(['Average Price'],axis=1)
sp_df['Ups and downs'] = sp_df['Close'].diff()
sp_df['Close Pct Change'] = sp_df['Close'].pct_change() * 100
sp_df = sp_df.iloc[1:,:]
sp_df.columns = ['date','Opening price', 'Highest price', 'Lowest price','Closing price', 'Volume(share)', 'Turnover(USD)', 'Ups and downs', 'Change(%)']
sp_df = sp_df.set_index('date')
```

```python
sp_df.to_csv("SP500.csv", header=True, index=True)
```
